```python
import pickle
from pathlib import Path

data_path = Path('../outputs/outputs/trials_data/trials_dataset.pkl')
with open(data_path, 'rb') as f:
    data = pickle.load(f)

trials = data['all_trials']
print(f'Loaded {len(trials)} trials')
```

```python
# Look at a single trial
i = 0
t = trials[i]
print(''.join(t['tokens']))
```
